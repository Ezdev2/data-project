```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données depuis le fichier CSV
data = pd.read_csv("sujet_2023_08_15.csv")


# Créer un graphique en barres pour le taux d'achèvement de l'école primaire
plt.figure(figsize=(10, 5))
plt.bar(data['Année'], data['Taux d’achèvement de l’école primaire (%)'])
plt.xlabel('Année')
plt.ylabel("Taux d'achèvement de l'école primaire (%)")
plt.title("Taux d'achèvement de l'école primaire à Madagascar")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# Créer un graphique en ligne pour la croissance du PIB
plt.figure(figsize=(10, 5))
plt.plot(data['Année'], data['Croissance du PIB (%)'], marker='o')
plt.xlabel('Année')
plt.ylabel('Croissance du PIB (%)')
plt.title('Croissance du PIB à Madagascar au fil des années')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()



# Créer un graphique en nuage de points pour visualiser la corrélation entre croissance du PIB et taux d'achèvement de l'école primaire
plt.figure(figsize=(10, 6))
plt.scatter(data['Croissance du PIB (%)'], data['Taux d’achèvement de l’école primaire (%)'])
plt.xlabel('Croissance du PIB (%)')
plt.ylabel("Taux d'achèvement de l'école primaire (%)")
plt.title("Impact de l'économie sur la scolarisation à Madagascar")
plt.grid(True)
plt.tight_layout()

# Calculer la corrélation entre les deux variables
correlation = data['Croissance du PIB (%)'].corr(data['Taux d’achèvement de l’école primaire (%)'])
plt.annotate(f"Corrélation = {correlation:.2f}", xy=(0.1, 0.9), xycoords='axes fraction', fontsize=12)

plt.show()



# Créer un graphique de tendances temporelles pour le taux d'achèvement de l'école primaire et la croissance du PIB
plt.figure(figsize=(10, 6))
plt.plot(data['Année'], data['Taux d’achèvement de l’école primaire (%)'], label="Taux d'achèvement de l'école primaire", marker='o')
plt.plot(data['Année'], data['Croissance du PIB (%)'], label="Croissance du PIB", marker='x')
plt.xlabel('Année')
plt.ylabel('Taux / Croissance (%)')
plt.title("Taux d'achèvement de l'école primaire et Croissance du PIB à Madagascar")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

```
